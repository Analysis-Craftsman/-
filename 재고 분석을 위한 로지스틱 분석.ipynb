```python
import pandas as pd

a_df = pd.read_excel("excel.xlsx", "기존 재고 데이터")
b_df = pd.read_excel("excel.xlsx", "수정된 재고 데이터")
```

```python
a_df
```

```python
b_df
```

```python
del b_df['재고']
b_df
```

```python
from sklearn.linear_model import LogisticRegression as logreg
from sklearn import metrics

reg = logreg()

x = b_df[b_df.columns.to_list()[:-1]]
y = b_df[b_df.columns.to_list()[-1]]

reg.fit( x, y )

predict = reg.predict( x )

score = metrics.accuracy_score( y, predict )
print( "정답률 판별 결과 :", round(score*100, 2), "%" )

report = metrics.classification_report( y, predict )
print( report )
```

```python
# 예측 기간 : 2020.10
# 마케팅팀과 협의를 통한 반드시 500 명 이상의 유입자를 만들 수 있도록 비용을 소비하기로 함 
# 사용자당 쿠폰은 2개 정도만 발급

import numpy as np

predict = reg.predict( np.array([2020.10, 500, 2]).reshape(1, -1) )
if predict == 0 :
    print( "정답률 : 70% / 2020.10, 재고는 남지 않을 것으로 확인됩니다." )
elif predict == 1 :
    print( "정답률 : 70% / 2020.10, 재고는 남을 것으로 확인됩니다." )
```
